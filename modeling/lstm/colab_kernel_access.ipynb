Shell script for accessing GPU kernel on Google Colab through the Colab extension on VS Code.  Connect to a kernel and clone your repository.  

- `cd modeling/lstm`  
- `pip install -r colab_requirements.txt`  
- `export PYTHONPATH=/content/ADS599-Capstone:$PYTHONPATH` to tell python where to find the modules
- `python -m modeling.lstm.train`  

Peaks around 70 gb of RAM even with down-casting so add a filter to remove data if you don't have that much RAM
